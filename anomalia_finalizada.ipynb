```
import numpy

valores = [8, 9, 11, 12, 13, 16, 10, 5, 9, 12, 11, 8, 9, 11, 12, 15]
print("Para os valores: {}.".format(valores))
valores.sort()
print("Colocando-os em ordem, temos: {}.".format(valores))

print("Afim de encontrar os considerados anômalos, é necessário analisar os dados fornecidos e descobrir os limites que irão definir as anomalias em uma lista de dados,")

#encontrar média:
print("-------------------------")
soma = numpy.sum(valores)
media = soma / len(valores)
print("A média dos números fornecidos é :", media)


#encontrar os quartis:
print("-------------------------")
print("Quantidade de valores: {}".format(len(valores)))    #como são 16 valores, chega-se a conclusão de que os números na posição 1/4(valores[4]) e 3/4(valores[12]) são os quartis.
print("Quartil 1 =", valores[4])  #1/4
print("Quartil 2 =", valores[12])  #3/4
print("-------------------------")

#encontrar amplitude:
ampl = valores[12] - valores[4]
print("Amplitude =", ampl)
print("-------------------------")

#encontrar limites:
lim1 = media - 1.5 * ampl
lim2 = media + 1.5 * ampl

print("O primeiro limite é: ", lim1)
print("O o segundo limite é:", lim2)
print("-------------------------")

print("Neste caso os limites são para valores abaixo de 6,18 e acima de 15,18.")

for valor in valores:
  anom1 = valores[0]
  anom2 = valores[-1]

print("Ou seja, os valores considerados anomalos em tal lista são: {} e {}.".format(anom1, anom2))
print("A lista de valores anomalos é:")
print("Anomalias = [5, 16].")
```
