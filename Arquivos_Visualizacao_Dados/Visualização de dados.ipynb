```python
import numpy as np
import pandas as pd

iris = pd.read_csv('iris.csv')
```

```python
iris.boxplot(figsize=(15,7))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 1) existe alguma relação entre as classes confrontando seus atributos?
#    sim, alguns atributos separam as classes

# 2) a iris-setosa ela está bem separada das outras classes?
#    sim, não tem sobreposição

sns.pairplot(iris, hue="classe", height=3.5)
plt.show()
```

```python
# Iris-setosas -> largura_petala < 0.7 E comprimento_petala < 2.5

from collections import Counter

resultados = iris['classe'][(iris['largura_petala'] < 0.7) & ( iris['comprimento_petala'] < 2.5)]
print(Counter(resultados))
```

```python
iris['predicao'] = ''
iris
```

```python
iris['predicao'][(iris['largura_petala'] < 0.7) & ( iris['comprimento_petala'] < 2.5)] = 'Iris-setosa'
iris['predicao'][(iris['comprimento_petala'] > 2.4)  &  (iris['comprimento_petala'] < 4.8)] = 'Iris-versicolor'
iris['predicao'][(iris['largura_petala'] <= 3) & ( iris['comprimento_petala'] > 5)] = 'Iris-virginica'
iris
```

```python
(iris['classe'] == iris['predicao']).sum() / len(iris)
```

### Correlação

```python
iris['classe'][iris['classe'] == 'Iris-setosa'] = 0
iris['classe'][iris['classe'] == 'Iris-versicolor'] = 1
iris['classe'][iris['classe'] == 'Iris-virginica'] = 2

iris['classe'] = pd.to_numeric(iris['classe'])

iris.corr()
```

### Titanic

```python
titanic = pd.read_csv('titanic.data')

titanic
```

```python
titanic.info()
```

```python
titanic.isna().sum()
```

```python
## tratamento para dados ausentes ou faltantes
titanic.fillna({'Age':titanic.Age.mean()}, inplace=True)

## tratamento para dados categóricos
titanic['Sex'][titanic['Sex'] == 'male'] = 0
titanic['Sex'][titanic['Sex'] == 'female'] = 1
titanic['Sex'] = pd.to_numeric(titanic['Sex'])
```

```python
titanic.isna().sum()
```

```python
g = sns.FacetGrid(titanic, col="Survived")
g.map(plt.hist, 'Age', bins=20)
```

```python
g = sns.FacetGrid(titanic, col="Survived", row="Pclass", height=2.2)
g.map(plt.hist, 'Age', bins=20)
g.add_legend()
```
